Чтобы проверить, одинаковы ли размеры гранул в обучающей и тестовой выборках,сформулируем и проверим следующие статистические гипотезы:

- нулевая гипотеза H₀: средний размер гранул обучающей выборки равен серднему размеру гранул тестовой выборки;

- альтернативная гипотеза H₁: редний размер гранул обучающей выборки не равен серднему размеру гранул тестовой выборки;

Установим уровень значимости alpha = 0.01 Если p-value окажется меньше него - отвергнем гипотезу H₀ Установим параметр equal_var = False. Используем t-тест Стьюдента для проверки гипотез.

```python
state = np.random.RandomState(12345)
train_primary_cleaner_size_research = train['primary_cleaner.input.feed_size'].dropna().sample(5856, replace=True, random_state=state)
train_primary_cleaner_size_research.describe()
```

```python
test_primary_cleaner_size_research = test['primary_cleaner.input.feed_size'].dropna()
test_primary_cleaner_size_research.describe()
```

```python
alpha = .01
results = st.ttest_ind(train_primary_cleaner_size_research, test_primary_cleaner_size_research, equal_var = False)
print('p-значение: {:.40f}'.format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

***Вывод:***

Визуально распределения размера гранул как на этапе флотации, так и на этапе первичной очистки, похожи. Что подтверждается и ститистическим тестом. Следовательно, можно использовать эти выборки при моделировании.

## Модель

### Функция для вычисления итоговой sMAPE

Напишем функцию для вычисления симметричного среднего абсолютного процентного отклонения, по которому будем в дальнейшем определять лучшую модель.

```python
def smape(target, predict):
    result = (abs(target - predict) / ((abs(target) +abs(predict)) / 2)) * 100
    mean = result.mean()
    return mean
```

Напишем функцию для расчёта итоговой sMAPE

```python
def SMAPE_SUMMARY(smape_r, smape_f):
    answer = 0.25 * smape_r + 0.75 * smape_f
    return answer
```
